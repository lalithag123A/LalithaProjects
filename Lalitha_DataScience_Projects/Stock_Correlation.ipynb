In terms of investing, it is not a good idea to buy different stocks that are correlated. Owning stocks 
that are not correlated reduces the risk to oneâ€™s portfolio. 
For this task, you want to find the correlation between each of the following stocks BITCOIN, GOOG, 
MSFT and AAPL. For the correation, you will compare two at a time. There will involve 6 pairwise 
comparisions (BITCOIN and GOOG; BITCOIN and MSFT; BITCOIN and AAPL; GOOG and MSFT; GOOG 
and AAPL; MSFT and AAPL).
Performing a correlation assumes there are an equal number of samples in the BITCOIN, GOOG, MSFT 
and AAPL data sets. Programmatically determine which data set has the fewest number of rows and 
pare down the other data sets. When paring down a data set, make sure to remove the oldest records.
Attached is EXCEL file called Correlation-between-Two-Stocks. It gives an example of finding the 
correlation between 2 stocks. The Manual Formula tab shows the details for finding the correlation 
between 2 stocks. Use this tab as a guide to find the correlation between 2 stocks. Make this into a 
function that may be called 6 times. You will want to pass in 2 stocks and return the correlation value

```python
# pd.read_csv reads all the csv files
dfb = pd.read_csv("BITCOIN.csv")
dfg = pd.read_csv("GOOG.csv")
dfm = pd.read_csv("MSFT.csv")
dfa = pd.read_csv("AAPL.csv")

def stockcorr(df1, df2): 
    df1 = df1.sort_values(by="Date", ascending=False)
    df2 = df2.sort_values(by="Date", ascending=False)
    #get matched rows by the date and ignore rest
    dfcom = pd.merge(df1, df2, how="inner", on = "Date") 
    #find correlation between 2 stocks 
    corr = dfcom['Close_x'].corr(dfcom['Close_y'])    
    return corr

stocksTitle = ["BITCOIN and GOOG", "BITCOIN and MSFT", "BITCOIN and AAPL", "GOOG and MSFT", "GOOG and AAPL", "MSFT and AAPL"]
    
for stock in stocksTitle:
    if (stock == "BITCOIN and GOOG"):
        print("BITCOIN and GOOG Stock Corrilation: "+ str(stockcorr(dfb, dfg)))
    if (stock == "BITCOIN and MSFT"):
        print("BITCOIN and MSFT Stock Corrilation: "+ str(stockcorr(dfb, dfm)))        
    if (stock == "BITCOIN and AAPL"):
        print("BITCOIN and AAPL Stock Corrilation: "+ str(stockcorr(dfb, dfa)))        
    if (stock == "GOOG and MSFT"):
        print("GOOG and MSFT Stock Corrilation: "+ str(stockcorr(dfg, dfm)))
    if (stock == "GOOG and AAPL"):
        print("GOOG and AAPL Stock Corrilation: "+ str(stockcorr(dfg, dfa)))
    if (stock == "MSFT and AAPL"):
        print("MSFT and AAPL Stock Corrilation: "+ str(stockcorr(dfm, dfa)))
```
